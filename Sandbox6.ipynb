```python
test = np.random.uniform(20, size=20)
```

```python
df_test = pd.DataFrame(test, columns=["Q"])
df_test
```

```python
Q = "Q"
nan = [(3,6,4,3), (12,18,7,5)]
df_test["Qb"] = pd.Series(dtype=np.float64)
df_test
```

```python
s = 0
# Does all but tail
for tup in nan:
    v1 = df_test[Q].iloc[s:tup[0]].to_numpy()*2
    df_test["Qb"].iloc[s:tup[0]] = v1
    s = tup[1] + 1
# Tail
r = 2
 
c = len(df_test["Q"].iloc[s:]) >= r # This works to check if the end is valid(This should always be the case)

if c:
    v1 = df_test[Q].iloc[s:].to_numpy()*2
    df_test["Qb"].iloc[s:] = v1
else:
    df_test["Qb"].iloc[s:] = np.full(len(df_test["Qb"].iloc[s:]), -9999)


df_test
```
